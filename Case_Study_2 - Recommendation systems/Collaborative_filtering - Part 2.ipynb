## **Project Overview**

Objective: The goal of this project is <u>to build a personalized book recommendation system that suggests books to users based on their preferences and the preferences of similar users</u>. Specifically, we aim to answer the question: *"What other books would you recommend to someone who likes book X?".* 

Methods used: 
1. **Collaborative Filtering** - recommends items to users based on the behaviors and preferences of similar users. For instance, we can find users who have rated the same books highly as a target user and then suggest other books that these similar users enjoyed.
   
2. **Singular Value Decomposition (SVD)**  (particularly Truncated SVD, which is a common variation).
   SVD is a matrix factorization technique that decomposes the user-item matrix into three components: U (user features), Σ (singular values), and Vᵀ (item features). Unlike PCA, other matrix factorization techniques, which is mainly used for feature extraction and visualization, SVD is more appropriate for building recommender systems, natural language processing, and image compression.

The input data for the recommendation model is a <u>user-item matrix</u> where:

- Rows represent users.
- Columns represent items (books).
- Values are numeric ratings from 1 to 10, with missing ratings replaced by 0 to indicate no preference or opinion.
The user-item matrix must be free of NaN values, and all data should be numeric to ensure proper functioning of the model.


To improve the quality and relevance of the recommendations, <u> we will also incorporate additional user demographic and geographic data </u>. Information such as location or age can provide context that enriches the model's understanding of a user's preferences. For example, younger readers may have different tastes than older readers or a user from a specific country may prefer books in their native language.

Book recommendation dataset you can find [here](https://www.kaggle.com/datasets/arashnic/book-recommendation-dataset).

#### More occurrences with fewer dimensions or more dimensions with sparser occurrences?

How to effectively reduce dimensions to provide a suitable input for a model? I am considering two approaches:

1) Strict Filtering: Retain only books that have been rated by more than 50 users and users who have rated more than 100 books. This approach results in a matrix with dimensions 144 users by 1300 books.

2) Broader Inclusion: Expand the criteria to include books rated by at least 25 users and users who have rated more than 25 books. This approach increases the matrix dimensions to 1399 users by 3250 books. However, this will also introduce more zeros in the matrix, where zero indicates no rating or no opinion.

##### **Option 1: More occurrences with fewer dimensions** (144 x 1300)

Pros: The matrix is denser with more ratings available, leading to potentially <u>more accurate and reliable analysis</u>.With more occurrences, the data you work with is likely more meaningful, as <u>it reflects more user interactions</u>.

Cons: Reducing the number of dimensions may result in loss of detailed information about less frequently rated books and users.

##### **Option 2: More dimensions with sparser occurrences** (1399 x 3250)

Pros: <u> Captures more detailed and nuanced interactions </u> across a wider range of books and users.

Cons:  The matrix becomes much sparser, which can reduce the effectiveness of many analytical techniques and algorithms.

#### Ultimate choice

**My final choice** is to opt for a denser matrix with higher-quality data, which will lead to more accurate and meaningful analysis. Therefore, **I will choose Option 1**: retaining only books that have been rated by more than 50 users and users who have rated more than 100 books.

## **4. Singular Value Decomposition (SVD)** <a id="chapter4"></a>

### Choosing the number of components using the explained variance ratio

We have 1300 columns and normally I will set the initial number of components to 1300, however TruncatedSVD can only produce as many components as the minimum of the number of rows or columns in the matrix, so in our case 144, then let's set the initial number to 144.

```python
svd_144 = TruncatedSVD(n_components = 144)
U144 = svd_144.fit_transform(customer_item_matrix)
cum_exp_144 = svd_144.explained_variance_ratio_
#cum_exp_144
```

```python
U144.shape
```

```python
np.cumsum(cum_exp_144)
```

```python
cum_evr144 = list(zip(range(1,U144.shape[1]+1),np.cumsum(cum_exp_144)))
cum_evr144[:20]
```

```python
cum_evr144_df = pd.DataFrame(cum_evr144).set_index(0)
cum_evr144_df
```

```python
cum_evr144_df.plot(title ='Nr of components vs. cumulative explained variance ratio',xlabel ='number of components', ylabel = 'cumulative explained variance ratio',legend = False, color = '#1098ad');
```

With 80 components capturing approximately 88% of the explained variance in the model, we can now return to the code and adjust the number of components to 80.

```python
def filter_and_merge_books(user_isbn_rating, data):
    filtered_data = (
        data[data['isbn'].isin(user_isbn_rating['isbn'])]
        .drop_duplicates(subset='isbn', keep='first')
        .reset_index(drop=True)
        [['isbn', 'book_title', 'book_author', 'year_of_publication', 'publisher', 'language', 'category']]
    )
    
    # Merge the filtered data with the user's rated books and sort by rating
    result = (
        pd.merge(user_isbn_rating, filtered_data, on='isbn')
        .sort_values(by='rating', ascending=False)
    )
    
    return result
```

### Functions <a id="section_6_2"></a>

#### **Code**

```python
def create_new_user_df(customer_item_matrix, ratings, positions):
    # example
    # ratings = [5,2,7,9,1,3,5,3,7,7]
    # position_to_populate = [0,157,200, 278, 440, 600, 829, 990, 997, 1000]
    
    vector_length = customer_item_matrix.shape[1]
    new_user = np.full(vector_length, np.NaN)

    # Populate the new user vector with the provided ratings
    new_user[positions] = ratings

    # Create a DataFrame for the new user
    new_user_df = pd.DataFrame([new_user], columns=customer_item_matrix.columns)

    # Extract books and ratings from the new user DataFrame
    user_books_rated = new_user_df.T.dropna()
    user_books_rated = user_books_rated.reset_index()
    user_books_rated.columns = ['isbn', 'rating']

    return user_books_rated,new_user_df
```

```python
def filter_and_merge_books(user_isbn_rating, data):
    filtered_data = (
        data[data['isbn'].isin(user_isbn_rating['isbn'])]
        .drop_duplicates(subset='isbn', keep='first')
        .reset_index(drop=True)
        [['isbn', 'book_title', 'book_author', 'year_of_publication', 'publisher', 'language', 'category']]
    )
    
    # Merge the filtered data with the user's rated books and sort by rating
    result = (
        pd.merge(user_isbn_rating, filtered_data, on='isbn')
        .sort_values(by='rating', ascending=False)
    )
    
    return result
```

```python
def recommend_books_for_new_user(new_user_df, svd_model, customer_item_matrix, number=10):
    # Transform the new user's data using the trained SVD model
    new_user_u = svd_model.transform(new_user_df.fillna(0))
    
    # Obtain the components matrix from the SVD model
    VT = svd_model.components_
    
    # Calculate the predicted ratings by multiplying the transformed user data with the components matrix
    new_user_pred = np.dot(new_user_u, VT)  #(1 x 80) * (80 x 1300) = (1 x 1300)
    
    # Create a DataFrame with the predicted ratings
    recs = pd.DataFrame(new_user_pred, columns=customer_item_matrix.columns)
    
    # Sort the recommended books by predicted rating and get the top 10
    new_user_rec = recs.T.sort_values(by=0, ascending=False)[:number].reset_index()
    new_user_rec.columns = ['isbn', 'rating']
    
    return new_user_rec
```
